```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('co2.csv')

# Display first 5 rows
display(df.head())
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data =df,x='Engine Size(L)', y='CO2 Emissions(g/km)')
plt.title("Engine Size vs CO2 Emissions")
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emissions (g/km)")
plt.show()
```

```python
# Linear Regression using Gradient Descent (from scratch)

# Select features
x = df['Engine Size(L)'].values
y = df['CO2 Emissions(g/km)'].values

# Normalize data (important for stable learning)
x = (x - np.mean(x)) / np.std(x)
y = (y - np.mean(y)) / np.std(y)

# Initialize parameters
m = 0   # slope
b = 0   # intercept

alpha = 0.01
iteration = 1000
n = len(x)

# Training loop
for i in range(iteration):
    
    # Predicted values
    y_pred = m * x + b
    
    # Compute gradients (partial derivatives)
    dm = (-2/n) * np.sum(x * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    # Update parameters
    m = m - alpha * dm
    b = b - alpha * db

# Gradient Descent Training Completed 
print(f"Slope (m): {m:.4f}")
print(f"Intercept (b): {b:.4f}")
```

```python
# Plot normalized data with regression line

plt.figure(figsize=(6,4))

# Scatter plot
plt.scatter(x, y, label="Data Points")

# Regression line
plt.plot(x, m*x + b, color='red', label="Best Fit Line")

plt.title("Linear Regression using Gradient Descent")
plt.xlabel("Engine Size (Normalized)")
plt.ylabel("CO2 Emissions (Normalized)")
plt.legend()
plt.show()

print("Insight: The regression line shows a strong positive relationship between engine size and emissions.")
```

```python
# Evaluate model using R² score

correlation = np.corrcoef(x, y)[0,1]
r2 = correlation**2

print("=== Model Evaluation ===")
print("R² Score: %.2f"% r2)

print("Interpretation: R² close to 1 indicates a strong linear relationship.")
```
